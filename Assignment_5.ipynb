```python
# Task 1

import math

class Circle:

    def __init__(self,circle_radius):
        self.__circle_radius = circle_radius

    def getRadius(self):
        return self.__circle_radius
    
    def area(self):
        circle_area = math.pi * self.__circle_radius ** 2
        return circle_area

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())
c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
# Task 2

class Triangle:

    def __init__(self,base,height):
        self.__base = base
        self.__height = height

    def getBase(self):
        return self.__base
    
    def getHeight(self):
        return self.__height
    
    def area(self):
        triangle_area = 0.5 * self.__base * self.__height
        return triangle_area

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
# Task 5

class Student:

    def __init__(self,student_name,student_ID,student_department,email = None,password = None,login_status = None):
        self.student_name = student_name
        self.student_ID = student_ID
        self.student_department = student_department
        self.email = email
        self.password = password
        self.login_status = login_status
        self.temp = ""
        print("Student object is created!")

class Usis:

    def __init__(self):
        print("USIS is ready to use!")

    def login(self,student_class_1):
        
        if student_class_1.email is None and student_class_1.password is None:
            print("Email and password need to be set.")

        elif student_class_1.email is None:
            print("Email needs to be set.")

        elif student_class_1.password is None:
            print("Password needs to be set.")

        else:
            student_class_1.login_status = True
            print("Login successful!")

    def advising(self,student_class_2,*course_list):

        if student_class_2.login_status is None:
            print("Please login to advise courses!")

        elif student_class_2.login_status is True and len(course_list) == 0:
            print("You haven't selected any courses.")

        elif student_class_2.login_status is True and len(course_list) > 3:
            print("You need special approval to take more than 3 courses")

        else:
            print("Advising successful!")
            self.__final_temp = student_class_2.temp

            for subjects in course_list:
                self.__final_temp += subjects + ", "

            self.__final_temp = self.__final_temp[:len(self.__final_temp) - 2]

    def individualDetails(self,student_class_3):
        return f"Name: {student_class_3.student_name}\nID: {student_class_3.student_department}\nDepartment: {student_class_3.student_department}\nAdvised courses: {self.__final_temp}"

rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110","CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
# Task 6

class Train:

    def __init__(self,train_name,*stations):
        self.train_name = train_name
        self.stations = stations
        self.location_list = []
        self.detail_dict = {}

    def addPassenger(self,*passenger_class):
        self.passenger_class = passenger_class

        for items in self.passenger_class:

            print(f"{items.passenger_name} welcome aboard")

            if items.start is None:
                first_location = self.stations[0]

            else:
                first_location = items.start

            if items.stop is None:
                last_location = self.stations[-1]

            else:
                last_location = items.stop

            first_location_index = self.stations.index(first_location)
            last_location_index = self.stations.index(last_location)
            
            for another_items in range(first_location_index,last_location_index):
                items.count += 1

            items.fare += items.count * 100
            self.location_list.append(first_location)
            self.location_list.append(last_location)
            self.location_list.append(items.fare)
            self.detail_dict[items.passenger_name] = self.location_list
            self.location_list = []

    def allPassengerDetails(self):
        
        for items_1,items_2 in self.detail_dict.items():

            print(f"Name: {items_1},Start: {items_2[0]},Destination:\n{items_2[1]},Fair: ${items_2[2]}")

class Passenger:

    def __init__(self,passenger_name,start = None,stop = None):
        self.passenger_name = passenger_name
        self.start = start
        self.stop = stop
        self.count = 0
        self.fare = 0

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
# Task 7

class BracuStudent:

    def __init__(self,student_name,home,bus_pass = False):
        self.student_name = student_name
        self.home = home
        self.bus_pass = bus_pass

    def show_details(self):
        print(f"Student name: {self.student_name}\nLives in {self.home}\nHave Bus Pass? {self.bus_pass}")

    def get_pass(self):
        self.bus_pass = True

class BracuBus:

    def __init__(self,bus_location,max_passenger = 2):
        self.bus_location = bus_location
        self.max_passenger = max_passenger
        self.passenger_list = []
        self.count = 0

    def show_details(self):
        print(f"Bus Route: {self.bus_location}")
        print(f"Passengers Count: {len(self.passenger_list)} (Max: {self.max_passenger})")
        print(f"Passenger on board: {self.passenger_list}")

    def board(self,*boarded_passengers):
        
        if len(boarded_passengers) == 0:
            print("No passenger!")

        else:
            
            for names in boarded_passengers:

                if names.home != self.bus_location:
                    print("You got on wrong bus!")

                if names.bus_pass is False:
                    print("You don't have bus pass!")

                if names.home == self.bus_location and names.bus_pass is True:

                    if self.count == self.max_passenger:
                        print("Bus is full!")

                    else:

                        print(f"{names.student_name} boarded the bus")
                        self.passenger_list.append(names.student_name)
                        self.count += 1

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
# Task 8

class Library:

    def __init__(self,library_location,book_dict):
        self.library_location = library_location
        self.book_dict = book_dict
        self.borrower = {}

    def details(self):
        print(f"{self.library_location} Library Details")
        print(f"Borrower Details: {self.borrower}")
        print(f"Books availability\n{self.book_dict}")

class Reader:

    def __init__(self,name):
        self.name = name
        self.count = 0
        self.personal_dict = {}

    def borrow(self,library_class,*book_list):

        for books in book_list:

            for item_1,item_2 in library_class.book_dict.items():

                if books == item_1:

                    if item_2 > 0:

                        if self.count < 5:

                            self.count += 1
                            library_class.borrower[self.name] = self.count
                            library_class.book_dict[item_1] -= 1
                            print(f"{item_1} book is borrowed successfully.")

                            if item_1 not in self.personal_dict:
                                self.personal_dict[item_1] = 1

                            else:
                                self.personal_dict[item_1] += 1

                        else:
                            print("You cannot borrow more than 5 books.")

                    else:
                        print(f"{item_1} books are not available at the moment.")

    def readerInfo(self,*another_book_detail):
        
        if len(another_book_detail) == 0:

            print(f"{self.name}, you have {self.count} book(s) with you.")

            for item_1,item_2 in self.personal_dict.items():
                print(f"Books on {item_1}: {item_2}")

        else:

            for another_items in another_book_detail:

                for item_1,item_2 in self.personal_dict.items():
                    
                    if another_items == item_1:

                        print(f"{self.name}, you have {item_2} {item_1} book(s) with you.")

L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
